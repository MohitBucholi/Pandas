### Series in Pandas
A pandas series is like a columnin a table. It is one dimensional array holding data of any type

```python
# Crete a simple pandas series

# import pandas
import pandas as pd
a = [15,13,13,41]
x = pd.Series(a)
x
# we got two columns the 1st one is index that start from 0. and another one is elements.
```

```python
# access elements with the help of index
x[2]
```

#### Lable:
The lable is used to access a specified value. use lable insetied of index.

```python
a = [15,13,13,41]
x = pd.Series(a, index=['a', 'b','c','d'])
x
```

```python
# access element using lable
x['d']
```

Create a simple pandas series from dictionary:

```python
b =  {'day1':420,'day2':411,'day3':560,'day4':356}
y =  pd.Series(b)
print(y)
y['day3']  # access using key. here key's are act as lable(index)
```

### Data Frames:
Data sets in pandas are usually in multi- dimensionl table, called dataframes.

```python
# create a simple dataframe:
data = {"calories":[450,411,356,400,394,510,498],"duration":[55,50,45,48,40,60,58]}
df =  pd.DataFrame(data)
df
```

#### Locate row :
dataframe_name.loc[ 'index' ]

```python
# access a single row
df.loc[0]
```

```python
# access multiple row
df.loc[2:5]
```

```python
# access single column
df["duration"]
```

```python
# access a particular part of data_frame
df.loc[2:5,"calories"]
```

```python
df.loc[2:5,"calories":]
```

#### Lable to dataframe:

```python
data = {"calories":[450,411,356,400,394,510,498],"duration":[55,50,45,48,40,60,58]}
df =  pd.DataFrame(data,index=['day1','day2','day3','day4','day5','day6','day7'])
df
```

```python
# access using lable
print(df.loc['day7'])
df.loc['day2':'day5']
```

#### Data cleaning:
1. Empty cells
2. Data in wrong format
3. wrong data
4. duplicates

```python
df = pd.read_csv(r"D:\HP\datasets kaggle\calories_burn.csv")
df
```

##### Empty cells can potentially give the wrong result. When you analyze data. since data sets are big & removing a big row will not have a big impact on the result

```python
# it romove all empty row from data
df1 = df.dropna()
df1
```

```python
# fill missing values with mean median mode
df['calories'].fillna(df['calories'].mean(),inplace = True)   # inplace is used to change in original data
df['duration'].fillna(df['duration'].mean(),inplace =True)
df
```

```python
#check the duplicates value in data.by using 'duplicated()' it return boolean value
df.duplicated()
```

```python
# drop the duplicates
df.drop_duplicates(inplace=True)   # inplace is used to change in orignal data
df
```
